```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from wordcloud import WordCloud
import itertools
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
```

```python
airbnb = pd.read_csv("clean_airbnb.csv")
```

# Common Words in Listing Weighed by Price

```python
words = ((airbnb['name'] + ' ')* airbnb['price']).tolist() # list of lists of words
words_merged = list(itertools.chain(words)) # list of lists of one string (all the words are concatenated)
```

```python
text = ''
for i in range(len(words_merged)):
    text += str(words_merged[i])
    text += ' '
```

```python
wordcloud = WordCloud(
    width = 800, height = 800,
    min_font_size = 20,
    background_color = 'white',
    collocations = False
).generate(text)

plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```
