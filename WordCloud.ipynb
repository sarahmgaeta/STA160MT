```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import itertools
from wordcloud import WordCloud
```

```python
airbnb = pd.read_csv("clean_airbnb.csv")
```

# Common Words in All Listings

```python
text = ''
for i in range(len(airbnb['name'])):
    text += str(airbnb['name'][i])
    text += ' '
```

```python
wordcloud = WordCloud(
    width = 800, height = 800, # 8x8?
    min_font_size = 20, # play with font size
    background_color = 'white', # white to match LaTeX file
    collocations = False # ignore bigrams
).generate(text)

plt.figure(figsize = (8, 8), facecolor = None) # image size and coloring
plt.imshow(wordcloud, interpolation = "bilinear")
plt.tight_layout(pad = 0) # reduce space
plt.axis("off")
plt.savefig('AllCommonWordcloud.jpg')
plt.show()
```

# Common Words in Listings Over 500 Dollars

```python
airbnb_over_500 = airbnb[airbnb['price'] > 500]

# Create a text blob
text = ''
for i in airbnb_over_500['name']:
    text += str(i)
    text += ' '
```

```python
wordcloud = WordCloud(
    width = 800, height = 800,
    min_font_size = 20,
    background_color = 'white',
    collocations = False
).generate(text)

plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud, interpolation = "bilinear")
plt.tight_layout(pad = 0)
plt.axis("off")
plt.savefig('ExpensiveCommonWordcloud.jpg')
plt.show()
```
